```python
import requests
import bs4
import re
import time
import random
from datetime import date
```

```python
today = date.today()
filename = today.strftime("Prague-Bezrealitky-%Y-%m-%d")
file = open(f'{filename}.csv',mode='w',encoding='utf-8')
file.write('listing_id,layout,condition,floor_area,price,fees,deposit,district,building_type,penb,furnishing,floor,balcony,terrace,cellar,loggia,parking,elevator,garage,heating,age,latitude,longitude\n');
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
current_page = 1
max_pages = 999

while current_page <= max_pages:
    
    url = f'https://www.bezrealitky.com/listings/offer-rent/flat/praha?&page={current_page}'
    time.sleep(random.randrange(8,12))
    response = requests.get(url,headers=headers)
    soup = bs4.BeautifulSoup(response.text,"lxml")

    max_pages = int(soup.select('.page-link.pagination__page')[-2].getText())
    listings_in_page = soup.find_all(class_='product__link')
    
    print(f'Scraping page {current_page} of {max_pages}')

    for apartment in listings_in_page:

        apartment_url = apartment.get('href')
        url2 = 'https://www.bezrealitky.com' + apartment_url
        time.sleep(random.randrange(4,7))
        
        
        response2 =''
        while response2 == '':
            try:
                response2 = requests.get(url2,headers=headers)
                break
            except:
                print("Connection refused by the server. Will try to reconnect after short wait.")
                time.sleep(90)
        
        soup2 = bs4.BeautifulSoup(response2.text,"lxml")

        listing_id = ''
        layout = ''
        floor_area = ''
        price = ''
        fees = ''
        deposit = ''
        district = ''
        building_type = ''
        penb = ''
        furnishing = ''
        floor = ''
        balcony = ''
        terrace = ''
        cellar = ''
        loggia = ''
        parking = ''
        elevator = ''
        garage = ''
        condition = ''
        heating = ''
        age = ''
        latitude = ''
        longitude = ''  
        
        try:
            listing_id = soup2.find("th", text="Listing ID:").find_next_sibling("td").text
        except:
            pass
        try:            
            layout = soup2.find("th", text="Layout:").find_next_sibling("td").text
        except:
            pass
        try:
            floor_area = soup2.find("th", text="Floor Space:").find_next_sibling("td").text.replace(' mÂ²','')
        except:
            pass
        try:
            price = soup2.find("th", text="Price:").find_next_sibling("td").text.replace('CZK ','').replace(',','')
        except:
            pass
        try:
            fees = soup2.find("th", text="Fees:").find_next_sibling("td").text.replace('CZK ','').replace(',','')
        except:
            pass
        try:
            deposit = soup2.find("th", text="Refundable security deposit:").find_next_sibling("td").text.replace('CZK ','').replace(',','')
        except:
            pass
        try:
            district = soup2.find("th", text="Metropolitan district:").find_next_sibling("td").text
        except:
            pass
        try:
            building_type = soup2.find("th", text="Building Type:").find_next_sibling("td").text
        except:
            pass
        try:
            penb = soup2.find("th", text="PENB:").find_next_sibling("td").text
        except:
            pass
        try:
            furnishing = soup2.find("th", text="Furnishing and Fittings:").find_next_sibling("td").text
        except:
            pass
        try:
            floor = soup2.find("th", text="Floor:").find_next_sibling("td").text
        except:
            pass
        try:
            balcony = soup2.find("th", text="Balcony:").find_next_sibling("td").text
        except:
            pass
        try:
            terrace = soup2.find("th", text="Terrace:").find_next_sibling("td").text
        except:
            pass
        try:
            cellar = soup2.find("th", text="Cellar:").find_next_sibling("td").text
        except:
            pass
        try:
            loggia = soup2.find("th", text="Loggia:").find_next_sibling("td").text
        except:
            pass
        try:
            parking = soup2.find("th", text="Parking:").find_next_sibling("td").text
        except:
            pass
        try:
            elevator = soup2.find("th", text="Lift:").find_next_sibling("td").text
        except:
            pass
        try:
            garage = soup2.find("th", text="Garage:").find_next_sibling("td").text
        except:
            pass
        try:
            condition = soup2.find("th", text="Condition:").find_next_sibling("td").text
        except:
            pass
        try:
            heating = soup2.find("th", text="Heating:").find_next_sibling("td").text
        except:
            pass
        try:
            age = soup2.find("th", text="Age:").find_next_sibling("td").text
        except:
            pass
        try:
            latitude = soup2.find_all("div", {"class": "b-map__inner"})[0]['data-lat']
        except:
            pass
        try:
            longitude = soup2.find_all("div", {"class": "b-map__inner"})[0]['data-lng']
        except:
            pass
        
  
        file.write(f'{listing_id},{layout},{condition},{floor_area},{price},{fees},{deposit},{district},{building_type},{penb},{furnishing},{floor},{balcony},{terrace},{cellar},{loggia},{parking},{elevator},{garage},{heating},{age},{latitude},{longitude}\n');
    
    current_page += 1 
    
file.close()
```
